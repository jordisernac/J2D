# Transformacións y Estandaritzacións

Per tal de facilitar les tasques de transformació del les dades dels nostres datasets, realitzarem les transformacions per separat.

## Transformacións de df_lloguer

```python
# imputem els nuls per 0
df_lloguer['Preu'].fillna(0, inplace=True)
```

```python
# Separem les dades de preu_mitja en dues columnes independents

import pandas as pd

def transform_lloguer_dataframe(df):
    
    df_filtered = df[(df['Lloguer_mitja'] == 'Lloguer mitjà per superfície (Euros/m2 mes)') |
                     (df['Lloguer_mitja'] == 'Lloguer mitjà mensual (Euros/mes)')]

    
    result_df = df_filtered.pivot(index=['Any', 'Trimestre', 'Codi_Districte', 'Nom_Districte', 'Codi_Barri', 'Nom_Barri'],
                                  columns='Lloguer_mitja',
                                  values='Preu')

    
    result_df.columns = ['Preu Lloguer mitja mensual (Euros/mes)', 'Preu Lloguer mitja per superficie (Euros/m2 mes)']
    result_df.reset_index(inplace=True)

    return result_df

df_lloguer_trans = transform_lloguer_dataframe(df_lloguer)

df_lloguer_trans.head()
```

```python

# Min-Max scaler per 'Preu Lloguer mitja mensual (Euros/mes)'
scaler_mensual = MinMaxScaler()
df_lloguer_trans['Preu Lloguer mitja mensual (Euros/mes)'] = scaler_mensual.fit_transform(df_lloguer_trans['Preu Lloguer mitja mensual (Euros/mes)'].values.reshape(-1, 1))

# Min-Max scaler per 'Preu Lloguer mitja per superficie (Euros/m2 mes)'
scaler_superficie = MinMaxScaler()
df_lloguer_trans['Preu Lloguer mitja per superficie (Euros/m2 mes)'] = scaler_superficie.fit_transform(df_lloguer_trans['Preu Lloguer mitja per superficie (Euros/m2 mes)'].values.reshape(-1, 1))

df_lloguer_trans.head()
```

```python
# Finalment despleguem els trimestres en columnes per facilitar la unio entre les dues taules
   
def pivot_trimestres(df):    
    df_lloguer_pivot_trimestres = df_lloguer_trans.pivot(index=['Codi_Districte', 'Codi_Barri'],
                                     columns='Trimestre',
                                     values=['Preu Lloguer mitja mensual (Euros/mes)', 'Preu Lloguer mitja per superficie (Euros/m2 mes)'])

    df_lloguer_pivot_trimestres.columns = [f'{col[0]}_T{col[1]}' for col in df_lloguer_pivot_trimestres.columns]
    
    return df_lloguer_pivot_trimestres

df_lloguer_pivot_trimestres = pivot_trimestres(df_lloguer_trans)
df_lloguer_pivot_trimestres.head()
```

## Transformacións del df_soroll

```python
def convertir_a_numerico(df):
    
    df['Valor'] = df['Valor'].str.rstrip('%').astype(float)

    
    df['Valor'] = df['Valor'] / 100

    return df


df_soroll = convertir_a_numerico(df_soroll)


df_soroll.head()
```

```python

def transform_soroll_dataframe(df_soroll):
    df_soroll_transformed = df_soroll.pivot(index=['Codi_Districte', 'Nom_Districte', 'Codi_Barri', 'Nom_Barri', 'Concepte'],
                                           columns='Rang_soroll',
                                           values='Valor')
        
    df_soroll_transformed.reset_index(inplace=True)
        
    df_soroll_transformed.columns.name = None
        
    return df_soroll_transformed

df_soroll_transformed = transform_soroll_dataframe(df_soroll)

df_soroll_transformed.head()
```

```python
df_soroll_transformed.info()
```

```python
# Finalment despleguem els conceptes en columnes i eliminem els noms dels districtes i dels barris per facilitar la unio entre les dues taules
   
def pivot_conceptes(df):    
    df_soroll_pivot_conceptes = df_soroll_transformed.pivot(index=['Codi_Districte', 'Codi_Barri'],
                                     columns='Concepte',
                                     values=['40-45 dB', '45-50 dB', '50-55 dB', '55-60 dB', '60-65 dB', '65-70 dB', '70-75 dB', '75-80 dB', '<40 dB', '>=80 dB' ])

    df_soroll_pivot_conceptes.columns = [f'{col[0]}_{col[1]}' for col in df_soroll_pivot_conceptes.columns]
    
    return df_soroll_pivot_conceptes

df_soroll_pivot_conceptes = pivot_conceptes(df_soroll_transformed)
df_soroll_pivot_conceptes.head()
```

```python
df_soroll_pivot_conceptes.info()
```

# Unió dels datasets

```python
df_final = df_lloguer_pivot_trimestres.merge(df_soroll_pivot_conceptes, how='left', on=['Codi_Districte', 'Codi_Barri'])
```

# Anàlisi de components

```python
from sklearn.decomposition import PCA
```

```python
# Crear una instancia de PCA amb el umbral de varianza del 95%
pca = PCA(n_components=0.95)

# Aplicar PCA a les dades
principal_components = pca.fit_transform(df_final)

# Núumero  componentes retingudes
n_components_retained = pca.n_components_

# Crear un nou DataFrame amb les components principals
df_pca = pd.DataFrame(data=principal_components, columns=[f'PC{i}' for i in range(1, n_components_retained + 1)])

# Observar la variansa explicada por cada component principal
explained_variance_ratio = pca.explained_variance_ratio_
print("Varianza explicada por cada componente principal:", explained_variance_ratio)

# Suma acumulada de la variansa explicada
cumulative_explained_variance = explained_variance_ratio.cumsum()
print("Varianza explicada acumulada:", cumulative_explained_variance)
```
